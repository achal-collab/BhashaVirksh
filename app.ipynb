```python
import streamlit as st
from deep_translator import GoogleTranslator

# 1. Page Configuration (The "App" Look)
st.set_page_config(page_title="IndicTrans Pro", page_icon="🌐")

st.title("🌐 Universal Indian Language Translator")
st.markdown("---")

# 2. Map Indian Languages to their codes
indian_languages = {
    "Hindi": "hindi",
    "Bengali": "bengali",
    "Gujarati": "gujarati",
    "Kannada": "kannada",
    "Malayalam": "malayalam",
    "Marathi": "marathi",
    "Punjabi": "punjabi",
    "Tamil": "tamil",
    "Telugu": "telugu",
    "Urdu": "urdu",
    "English": "english"
}

# 3. The Interface (App Sidebar and Layout)
col1, col2 = st.columns(2)

with col1:
    st.subheader("Input")
    source_text = st.text_area("Type your text here (Any language)...", height=200)

with col2:
    st.subheader("Settings")
    target_lang_name = st.selectbox("Select Target Language", list(indian_languages.keys()))
    target_code = indian_languages[target_lang_name]
    
    translate_button = st.button("Translate Now →", use_container_width=True)

# 4. The "Brain" (Logic)
if translate_button:
    if source_text.strip() == "":
        st.warning("Please enter some text first!")
    else:
        with st.spinner('Translating...'):
            try:
                # 'auto' detects the input language automatically
                translated = GoogleTranslator(source='auto', target=target_code).translate(source_text)
                
                st.markdown("---")
                st.success(f"**Translated Text ({target_lang_name}):**")
                st.markdown(f"### {translated}")
                
                # Bonus: Copy Button
                st.button("📋 Copy to Clipboard", on_click=lambda: st.write("Text Copied!"))
            except Exception as e:
                st.error(f"Error: {e}")

# 5. Footer for Students
st.markdown("---")
st.caption("Demo Platform created for Educational Purposes | Powered by Google Translate Engine")
```
